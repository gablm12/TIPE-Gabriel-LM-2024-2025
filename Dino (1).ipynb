II IA
Tentative d'IA (version du jeu sans le katana) basée sur des actions prenant en input des paramètres tels que des distances, afin de lui laisser la possibilité d'évoluer. Les paramètres (distances et choix des objets déclencheurs de chaque action) sont choisis aléatoirement, puis rapproché de la valeur moyenne du paramètre correspondant. Ces valeurs moyennes sont des moyennes définies pour chaque paramètre et se modifiant après chaque itération de la boucle d'entrainement, de manière pondérée par le score effectué lors de l'itération.
Le module renvoie par la suite la liste des scores et paramètres utilisés, puis le score maximal trouvé, ainsi que les paramètres associés à ce score.
(1000 itérations prennent environ 5 secondes)

```python
import random as r
#Itérations
n = 2000
#Création des variables:
coefTot, coef = 0, 0
dJm, dSm, dPm, dySm = 0, 0, 0, 0
cJm, cSm, cPm = 0, 0, 0
resultats_score = []
resultats_var = []
score_moy = 200
score_max, i_max = 800, 0
def conv(x):
    if x > 2/3:
        return "bat"
    elif x > 1/3:
        return "cactus"
    else:
        return "both"
#Définir des actions:
def jump(choixJ):
    global sautInitié
    if sautInitié == "non":
        if choixJ == "cactus" and procheCacJ:
            sautInitié = "montée"
        elif choixJ == "bat" and procheBatJ:
            sautInitié = "montée"
        elif choixJ == "both" and (procheCacJ or procheBatJ):
            sautInitié = "montée"
def useSneak(choixS):
    global sneak
    if (choixS == "cactus" and procheCacS) or (choixS == "bat" and procheBatS) or (choixS == "both" and (procheBatS or procheCacS)):
        sneak = True
def usePowerUp(choixP):
    if powerUp and sautInitié != "non":
        return ((choixP == "cactus" and procheCacP) or (choixP == "bat" and procheBatP) or (choixP == "both" and (procheBatP or procheCacP)))
#Faire n runs:
for i in range(n):
    #Affectation des paramètres
    xJ, xS, xP, xyS = 50 + 450*r.random(), 50 + 450*r.random(), 50 + 450*r.random(), 138*r.random()
    dJ, dS, dP, ySaut = xJ+ (dJm - xJ)*((r.random())**(1/2.4)), xS+ (dSm - xS)*((r.random())**(1/2.4)), xP+ (dPm - xP)*((r.random())**(1/2.4)), xyS + (dySm-xyS)*((r.random())**(1/2.4))
    xcJ, xcS, xcP = r.random(), r.random(), r.random()
    cJ, cS, cP = xcJ + (cJm - xcJ)*((r.random())**(1/2.4)), xcS + (cSm - xcS)*((r.random())**(1/2.4)), xcP + (cPm - xcP)*((r.random())**(1/2.4))
    choixJ, choixS, choixP = conv(cJ), conv(cS), conv(cP)
    #Réinitialisation des variables de jeu
    hautdécl, étapeSaut = 0, 0
    powerUp, power, power0 = False, 0, 0
    xcactus1, xcactus2, xcactus3, xcactus4, xcactus5, xvelo, yvelo, ydino = 2000, 2800, 300000, 400000, 400000, 2900, 250, 270
    sautInitié, étapeSaut = "non", 0
    cacspeed, alive = 3, True
    tick_survécus, score = 0, 0
    #Entrainement
    while alive==True:
        sneak = False
        procheCacJ = ( 30 < xcactus1 < 30 + dJ) or (30 < xcactus2 < 30 + dJ) or (30<xcactus2 < 30 + dJ) or (30<xcactus3 < 30 + dJ) or (30<xcactus4 < 30 + dJ)
        procheBatJ = 30 < xvelo < 30 + dJ
        procheCacS = ( 10 < xcactus1 < 30 + dS) or (30 < xcactus2 < 30 + dS) or (30<xcactus2 < 30 + dS) or (30<xcactus3 < 30 + dS) or (30<xcactus4 < 30 + dS)
        procheBatS = 10 < xvelo < 30 + dS
        procheCacP = ((270 - ydino) > ySaut) and (( 25 < xcactus1 < 30 + dP) or (25 < xcactus2 < 30 + dP) or (25 <xcactus2 < 30 + dP) or (25 <xcactus3 < 30 + dP) or (25 <xcactus4 < 30 + dP))
        procheBatP = ((270 - ydino) > ySaut) and (25 < xvelo < 30 + dP)
        tick_survécus += 1
        powerUp = True if power > 300 else False
        jump(choixJ)
        useSneak(choixS)
        if usePowerUp(choixP):
            sautInitié, étapeSaut = "montée", 0
            powerUp, hautdécl, power0 = False, (270-ydino), round(tick_survécus*(2.22/6))
        if sautInitié == "montée":
            ydino -= 17 * (11-étapeSaut)/8
            étapeSaut += 1
            if étapeSaut == 11:
                sautInitié = "descente"
                étapeSaut = 0
        if sautInitié == "descente":
            if hautdécl != 0:
                ydino += hautdécl/11
            étapeSaut += 1
            ydino += 17 * étapeSaut/8
        if étapeSaut == 11:
            if hautdécl != 0:
                hautdécl = 0
            sautInitié = "non"
            étapeSaut = 0
        xcactus1 -=(8*((cacspeed+1)**(0.42)))
        xcactus2 -=(8*((cacspeed+1)**(0.42)))
        xcactus3 -=(8*((cacspeed+1)**(0.42)))
        xcactus4 -=(8*((cacspeed+1)**(0.42)))
        xvelo -= (8*((cacspeed+1)**(0.5)))
        power = round(tick_survécus*(2.22/6))-power0
        if xcactus1 < -30:          
            xcactus1 = 2000
            if r.randint(0,1)==0:
                xcactus4= xcactus1 +50
            cacspeed += 0.5
        if xcactus2 < -30:
            xcactus2 = xcactus1+800
            if xcactus3 > 10000 and powerUp and r.randint(0,5)>=2 and xcactus2>2000:
                xcactus3, xcactus5 = xcactus2 + 150, xcactus2+50
        if xcactus3 < -30:
            xcactus3 = 100000
        if xcactus4 < -30:
            xcactus4 = 30000
        if xcactus5 < -30:
            xcactus5 = 30000
        if xvelo <-30 and xcactus2>2000:
            xvelo = xcactus2+r.randint(300, 400) if r.randint(0,1)==0 else xcactus2+r.randint(650, 850)
            a=r.randint(0,2)
            yvelo = 250 if a==0 else (210 if a==1 else (290))
        if (((10 < xcactus1 < 40) or (10 < xcactus2 < 40) or (10 < xcactus3 < 40) or (10 < xcactus4 < 40)) and (ydino > 190)) or ((not powerUp) and (10 < xvelo < 40) and ((yvelo-60 < ydino-10 < yvelo) or (((not sneak) and (yvelo-60 < ydino-60 < yvelo)) or (sneak and yvelo-60 < ydino-20 < yvelo)))) or ((powerUp) and (10 < xvelo < 40) and ((yvelo-60 < ydino+10 < yvelo) or (((not sneak) and (yvelo-60 < ydino-50 < yvelo)) or (sneak and yvelo-60 < ydino-20 < yvelo)))):
            alive = False
    #Ajuster les coefficients, stocker les résultats
    score = round(tick_survécus*(2.22/6))
    if score > score_max:
        score_max = score
        i_max = i
    coefSerie = (score/score_moy)**(3.5)
    resultats_score.append(score)
    resultats_var.append([dJ, dS, dP, ySaut, choixJ, choixS, choixP])
    score_moy = (score_moy*i+score)/(i+1)
    dJm = ((dJm*coefTot)+dJ*coefSerie*(i+1))/(coefTot+coefSerie*(i+1))
    dSm = ((dSm*coefTot)+dS*coefSerie*(i+1))/(coefTot+coefSerie*(i+1))
    dPm = ((dPm*coefTot)+dP*coefSerie*(i+1))/(coefTot+coefSerie*(i+1))
    dySm = ((dySm*coefTot)+ySaut*coefSerie*(i+1))/(coefTot+coefSerie*(i+1))
    cJm = ((cJm*coefTot)+cJ*coefSerie*(i+1))/(coefTot+coefSerie*(i+1))
    cSm = ((cSm*coefTot)+cS*coefSerie*(i+1))/(coefTot+coefSerie*(i+1))
    cPm = ((cPm*coefTot)+cP*coefSerie*(i+1))/(coefTot+coefSerie*(i+1))
    coefTot += coefSerie*(i+1)
    alive = True
#Récupération des résultats finaux
print(resultats_score)
print(resultats_var)
print(resultats_score[i_max], resultats_var[i_max])
```
